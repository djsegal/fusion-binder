```julia
R = 3

cur_dict = Dict(
    "Plasma" => Dict(
        "color" => :red,
        "r" => 0.33,
        "max" => 2*pi,
        "opacity" => 0.1,
    ),
    "Tokamak" => Dict(
        "color" => :blue,
        "r" => 1.0,
        "max" => 3*pi/2,
        "opacity" => 0.2
    )
)

for (cur_key, cur_value) in cur_dict
    cur_value["x"] = (t,p) -> ( R + cur_value["r"] * cos(t) ) * cos(p)
    cur_value["y"] = (t,p) -> ( R + cur_value["r"] * cos(t) ) * sin(p)
    cur_value["z"] = (t,p) -> ( 0 + cur_value["r"] * sin(t) ) * 1
end

plot()

for (cur_key, cur_value) in cur_dict
    r = cur_value["r"]

    pc = 100
    tc = 100
    # # @manipulate for pc = 10:10:140, tc = 10:10:140

    work_pp = collect(linspace(0,cur_value["max"], pc))
    work_tt = collect(linspace(0,2*pi, tc))

    tt = collect(zip(work_tt[1:end], [work_tt[2:end]...,work_tt[1]]))
    pp = collect(zip(work_pp[1:end], [work_pp[2:end]...,work_pp[1]]))


    xtp = cur_value["x"]
    ytp = cur_value["y"]
    ztp = cur_value["z"]

    xx = []
    yy = []
    zz = []
    for cur_t in work_tt[1:end-1]
        for cur_p in work_pp
            push!(xx, xtp(cur_t, cur_p))
            push!(yy, ytp(cur_t, cur_p))
            push!(zz, ztp(cur_t, cur_p))
        end
        push!(xx, NaN)
        push!(yy, NaN)
        push!(zz, NaN)
    end

    plot!(xx,yy,zz, fill=false, opacity=cur_value["opacity"], color=cur_value["color"], label="")

    xx = []
    yy = []
    zz = []
    for cur_p in work_pp
        for cur_t in work_tt
            push!(xx, xtp(cur_t, cur_p))
            push!(yy, ytp(cur_t, cur_p))
            push!(zz, ztp(cur_t, cur_p))
        end
    end
    push!(xx, NaN)
    push!(yy, NaN)
    push!(zz, NaN)
    plot!(xx,yy,zz, fill=false, opacity=cur_value["opacity"], color=cur_value["color"], label="")

    if !iszero( work_pp[end] % (2*pi) )
        xx = []
        yy = []
        zz = []
        for cur_p in [work_pp[1], work_pp[end]]
            for cur_t in work_tt
                push!(xx, xtp(cur_t, cur_p))
                push!(yy, ytp(cur_t, cur_p))
                push!(zz, ztp(cur_t, cur_p))
            end
            push!(xx, NaN)
            push!(yy, NaN)
            push!(zz, NaN)
        end
        plot!(xx,yy,zz, fill=false, width=2, opacity=0.3, color=cur_value["color"], label="")
    end
    plot!([0],[0],[0], opacity=0.6, label=cur_key, color=cur_value["color"], xticks=false,yticks=false,zticks=false, zlabel="Height")

end

title!("Plasma-Tokamak Configuration")
xlabel!("Length")
ylabel!("Width")

plot!(aspect_ratio=:equal)
```

```julia


function custom_plot(epsilon, kappa, delta, label)
    tmp_val_1 = 0.3 * ( 1 - delta ^ 2 )
    tmp_val_2 = ( -1 + 2 * delta + tmp_val_1 ) / 8

    cc = [
        -delta/2, 1 - tmp_val_2,
        +delta/2, 0 + tmp_val_2
    ]

    npts = 151
    aa = linspace(0, 2*pi, npts)

    xx = zeros(aa)

    for (cur_i, cur_c) in enumerate(cc)
        xx += cur_c * cos.( (cur_i-1) * aa )
    end

    yy = kappa * sin.(aa)

    rr = radius * ( 1 + epsilon * xx )
    zz = radius * ( 0 + epsilon * yy )

    if epsilon == g_epsilon && kappa == g_kappa && delta == g_delta
        cur_style = :dash
    else
        cur_style = :solid
    end
    
    plot!(
        [-rr..., NaN, +rr...], 
        [+zz..., NaN, +zz...],
        label=label,
        style=cur_style,
        aspect_ratio=:equal
    )

    xaxis!([-10,+10])
end
```

```julia
function itr(cur_value::AbstractFloat, cur_count=11)
    dif_value = cur_value * 0.7
    min_value = cur_value - dif_value
    max_value = cur_value + dif_value
    linspace(min_value, max_value, cur_count)
end
```
