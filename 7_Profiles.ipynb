```julia
npara(rho, nun, no) = (1+nun)*no* (1-rho^2)^nun

function npede(rho, no, nun, nsep, nped, rhoped, betan)
    if rho <= rhoped
        nn = 1 - (rho/rhoped)^betan
        nn ^= nun
        nn *=(no - nped)
        nn += nped
    else
        nn = (1-rho)/(1-rhoped)
        nn *=(nped-nsep)
        nn += nsep
    end
    
    nn
end
```

```julia
using Plots
```

```julia
using Interact
```

```julia
rhos = linspace(0,1)
@manipulate for no=linspace(1,5,5), nun=linspace(0.4,1.8,11), nsep
      plot(rhos, xpara.(rhos, nunu))
end
```

```julia
rhos = linspace(0,1,1000)

npede.(rhos, 1.014, 1.000, 0.2, 0.6780, 0.940, 2.000)

plot(
    rhos,
    npede.(rhos, 1.014, 1.000, 0.2, 0.6780, 0.940, 2.000)
)

plot!(
    rhos,
    npara.(rhos,1.000, 1.014)
)

ylims!(0,2)
```
